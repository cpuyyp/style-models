```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import json
```

```python
with open(f'{data_folder}/pastel/pastel_tasks2labels.json', 'r') as f:
    tasks2labels = json.load(f)
```

```python
repo_folder = os.environ["style_models_repo_folder"] if "style_models_repo_folder" in os.environ else None
data_folder = f"{repo_folder}/data" if repo_folder else '../../data'
```

```python
for split in ['train', 'test', 'valid']:
    df = pd.read_csv(f'{data_folder}/pastel/raw/raw_{split}_sentences.csv')

    plt.figure(figsize=(20,8))
    plt.suptitle(f'{split} set distribution', fontsize=20)
    for i,task in enumerate(tasks2labels):
        plt.subplot(2,4,i+1)
        df[task].value_counts().plot(kind='bar')
#         df[task].hist(bins=np.arange(-0.5, len(df[task].unique())+0.5, 1), rwidth=0.5)
        plt.xticks(rotation = 45, ha='right', fontsize=16)
    plt.show()
```

```python
df = pd.read_csv(f'{data_folder}/pastel/raw/raw_valid_sentences.csv')
for i,task in enumerate(tasks2labels):
    print(f'majority classifier val_acc for task {task}:\t{df[task].value_counts().max()/df[task].value_counts().sum()}')
```
